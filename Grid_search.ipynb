```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/kathanbhavsar/Desktop/Python,DS,ML/08-Linear-Regression-Models/Advertising.csv')
```

```python
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,15,100],'l1_ratio':[.1,.5,.7,.95,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_model,param_grid=param_grid,scoring='neg_mean_squared_error',cv=5,verbose=1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```
