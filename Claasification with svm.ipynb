```python
from svm_margin_plot import plot_svm_boundary
```

```python
from sklearn.svm import SVC
```

```python
y = df['Virus Present']
X = df.drop('Virus Present',axis = 1)
```

```python
model = SVC(kernel='linear',C=1000)
```

```python
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
## Exploring Parameter and kernel type
## In sklearn C is inversely proportional ie the higher the C the lesser parameter it will allow in the margin and vice versa
```

```python
model = SVC(kernel='linear',C=0.005)
```

```python
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma='scale')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel = 'poly',C=1,degree=8)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {"C":[0.01,0.1,1],'kernel':['linear','rbf']}
```

```python
grid = GridSearchCV(svm,param_grid=param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```
