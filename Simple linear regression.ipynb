```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/kathanbhavsar/Desktop/Python,DS,ML/08-Linear-Regression-Models/Advertising.csv')
```

```python
## Since simple linear regression deals with only one feature ew combine all spenrt columns into one
df['total_spent'] = df["TV"]+df['radio']+df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(df,x = 'total_spent', y = 'sales')
```

```python
sns.regplot(df,x = 'total_spent', y = 'sales')
```

```python
X = df['total_spent']
y = df['sales']
```

```python
# we have y = b1x+b0
#we use polyfit of numpy to calculate this
np.polyfit(X,y,deg=1) #This gibes us the value of b1 and b0
```

```python
# Now lets say we want to generate a line based on this b1 and b0 
potential_spent = np.linspace(0,500,100)
```

```python
potential_sales = 0.04868788*potential_spent + 4.24302822
```

```python
#now we plot the potential spent and potential sales
plt.plot(potential_spent,potential_sales)
```

```python
sns.scatterplot(df,x = 'total_spent',y = 'sales')
plt.plot(potential_spent,potential_sales)
```

```python
#Lets plot a polynomial model 
np.polyfit(X,y,3)
```

```python
x = np.linspace(0,500,100)
```

```python
pot_spent = 3.07615033e-07*x**3 + -1.89392449e-04*x**2 + 8.20886302e-02*x + 2.70495053e+00
```

```python
sns.scatterplot(df,x='total_spent',y= 'sales') ## This is a polynomial reg grapg 
plt.plot(x,pot_spent,color = 'red')
plt.show()
```
