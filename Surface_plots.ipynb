```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])
a,b = np.meshgrid(a,b)
```

```python
print(a)
```

```python
print(b)
```

```python
fig= plt.figure()
axes = plt.axes(projection = '3d')
axes.plot_surface(a,b,a**2+b**2,cmap = 'viridis')
```

```python
x = np.arange(-1,1,0.02)
y=x
x,y = np.meshgrid(x,y)
```

```python
fig = plt.figure(dpi=200)
axes = plt.axes(projection = '3d')
axes.plot_surface(x,y,x**2+y**2,cmap = 'Set2')
```

```python
fig = plt.figure()
axes = plt.axes(projection = '3d')
axes.contour(x,y,x**2+y**2,cmap = 'rainbow')
```
