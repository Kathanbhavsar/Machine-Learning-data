```python
stmt_one = '''
df['Price per person'] = df[['total_bill','size']].apply(lambda df: price_per_person(df['total_bill'],df['size']),axis=1)
'''
stmt_two = '''
df['Price per person'] = (np.vectorize(price_per_person)(df['total_bill'],df['size']))
'''
```
