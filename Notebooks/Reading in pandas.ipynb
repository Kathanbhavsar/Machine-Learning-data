```python
df = pd.read_csv('/Users/kathanbhavsar/Desktop/Python,DS,ML/03-Pandas/Sales_Funnel_CRM.csv')
```

```python
df
```

```python
## Lets say we want to know how many licences we have sold to each company.So to visualise it better we use pivot table
```

```python
pd.pivot(data = df,index = 'Company', columns='Product',values = "Licenses").fillna(0)
```

```python
pd.pivot_table(data=df,index='Company', aggfunc='sum', values = ['Licenses','Sale Price']) ## By pivot table
```

```python
# We can also use groupby method
df.groupby('Company').sum().drop('Account Number',axis=1)
```

```python
## We want to find the sale a account manager has done per contact
pd.pivot_table(df,index=['Account Manager','Contact'],columns='Product',aggfunc=[np.sum],fill_value=0)
```

```python
## We want to find the sale a account manager has done per contact
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=["Sale Price"],aggfunc=[np.sum],fill_value=0,margins=True)
```
