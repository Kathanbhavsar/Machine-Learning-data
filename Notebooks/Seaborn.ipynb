## Displot and Histplot

```python
sns.set(style = 'darkgrid')
sns.displot(x = 'salary', data = df, bins = 10, color='red',edgecolor = 'blue',lw = 10) ## We can use all the matplot attributes in the seaborn since seaborn is build on matplotlib
```

```python
sns.displot(x = 'salary', data = df, bins = 10, color='red',edgecolor = 'blue',lw = 10,kde = True,rug = True) # We can plot kde and rug plot on displot
```

```python
sns.histplot(x = 'salary', data = df, bins = 10, color='red',edgecolor = 'blue',lw = 10)
```

```python
df  = pd.read_csv('/Users/kathanbhavsar/Desktop/Python,DS,ML/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,8),dpi = 200)
sns.boxplot(data = df,y= 'math score',x = 'test preparation course')
```

```python
df.columns
```

```python
plt.figure(figsize = (10,4))
sns.boxplot(data = df,y= 'math score',x = 'parental level of education',hue = 'test preparation course',palette="Set2")
plt.legend(bbox_to_anchor = (1.05,.5))
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(data = df,y= 'math score',x = 'parental level of education',hue='gender') ## This shows the kde 
plt.legend(bbox_to_anchor = (1.05,0.5))
plt.show()
plt.figure(figsize=(10,4))
sns.violinplot(data = df,y= 'math score',x = 'parental level of education',hue='gender',split=True)
# If you are comparing with hue its better to use boxplot instead of the violin plot
plt.legend(bbox_to_anchor = (1.05,0.5))
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(data = df,y= 'math score',x = 'parental level of education',split=True,inner='quartile')
```

```python
df.columns
```

```python
plt.figure(figsize=(8,4),dpi =200)
sns.swarmplot(data = df,x = 'math score',y = 'gender',hue = 'test preparation course',size = 2,dodge = True)
plt.legend(bbox_to_anchor = (1.35,0.5))
```
