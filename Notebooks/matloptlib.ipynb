```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
```

```python
y = x*2
```

```python
plt.plot(x,y)
plt.title("String")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,9)
plt.ylim(0,18)
plt.show()
plt.savefig('firstplot.jpg')
```

```python
a = np.linspace(0,10,11)
b = a**4
x = np.arange(10)
y = x*2
```

```python
fig = plt.figure()
#Large Axis
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b) 
axes1.set_xlim(0,10)
axes1.set_ylim(0,10000)
axes1.set_xlabel('A')
axes1.set_ylabel("B")
axes1.set_title("Power of 4")
#Small Axis
axes2 = fig.add_axes([0.2,0.2,0.25,0.25])
axes2.plot(x,y)
axes2.set_xlim(0,5)
axes2.set_ylim(0,20)
axes2.set_xlabel('A')
axes2.set_ylabel("B")
axes2.set_title("Zoomed In ")
#Small Axis
plt.show()
```

## Subplots in Matplotlib

```python
a = np.linspace(0,10,11)
b = a**4
x = np.arange(10)
y = 2*x 
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(x,y)
fig.subplots_adjust(wspace = 1, hspace = .5)
fig.suptitle("Sabka Title") ## To Add a super title


fig.savefig('subplots.png')
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.show()
plt.tight_layout() ## Ensures that there is equal spacing
```
