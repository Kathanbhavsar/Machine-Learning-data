```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_data = sns.load_dataset('tips') 
```

```python
tips_data
```

```python
sns.barplot(x='sex', y='tip', data = tips_data)
```

## Mean of yaxis will be the height of bars in y axis

```python
tips_data[tips_data['sex']=="Female"]['tip'].mean()
```

```python
tips_data[tips_data['sex']=="Male"]['tip'].mean()
```

```python
sns.barplot(x='sex', y='tip',hue='time', data = tips_data, estimator=np.mean )
```

## Count Plot

```python
sns.countplot?
```

```python
sns.countplot(x="smoker", data=tips_data)
```

```python
sns.countplot(x="smoker",hue='sex',data=tips_data)
```

## Box Plot


```python
sns.boxplot(x='day', y='tip', hue='sex', data=tips_data)
```

## Violin plot

```python
sns.violinplot(x='day', y='tip', hue='sex', data=tips_data)
```

## Distribution Plot

```python
sns.distplot(tips_data['tip']) # kernel density distribution 
```

## KDE Plot (Kernal density estimation)

```python
sns.kdeplot(tips_data['total_bill'])
```

 ## Joint Plot

```python
plt.scatter(x='total_bill', y='tip', data=tips_data)
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='hex')
sns.jointplot(x='total_bill', y='tip',data = tips_data, kind='kde')
sns.jointplot(x='total_bill', y='tip',data = tips_data, kind='reg')
```

## Pair plot

```python
sns.pairplot(tips_data, hue = 'sex')
```

## Heatmap

```python
flights = sns.load_dataset('flights')
```

```python
pivot=flights.pivot_table(index = 'month', columns='year', values= 'passengers')
```

```python
pivot
#To find the visualization maximum number of people flew across through airoplane
```

```python
sns.heatmap(pivot, cmap = 'coolwarm')
```

```python
sns.heatmap(tips_data.corr(), cmap = 'coolwarm')
```
